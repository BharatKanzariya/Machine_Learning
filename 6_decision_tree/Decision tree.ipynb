```python
import pandas as pd
```

```python
df = pd.read_csv('salaries.csv')
df
```

## Data Cleaning

```python
x = df.drop('salary_more_then_100k',axis='columns')
x
```

```python
y = df[['salary_more_then_100k']]
y
```

## Make numaric data from string becuse machine only understand numbers and sklearn LabelEncoder method will make it easy for you.

```python
from sklearn.preprocessing import LabelEncoder
lable = LabelEncoder()
```

```python
x['company_num'] = lable.fit_transform(x['company'])
x['job_num'] = lable.fit_transform(x['job'])
x['degree_num'] = lable.fit_transform(x['degree'])
x
```

```python
x_final = x.drop(['company','job','degree'],axis='columns')
x_final
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(x_final,y)
```

```python
model.score(x_final,y)
```

```python
model.predict([[1,0,0]])
```
