```python
import pandas as pd
```

```python
df = pd.read_csv('HR_Data_Exer.csv')
df.head()
```

```python
d_salary = pd.get_dummies(df.salary)
d_salary
```

```python
d_last_evaluation = pd.get_dummies(df.last_evaluation)
d_last_evaluation
```

```python
d_time_spend_company = pd.get_dummies(df.time_spend_company)
d_time_spend_company
```

```python
d_satisfaction = pd.get_dummies(df.satisfaction_level,prefix='satisfaction')
d_satisfaction
```

```python
d_hours = pd.get_dummies(df.average_montly_hours,prefix='hours')
d_hours
```

```python
d_projects = pd.get_dummies(df.number_project)
d_projects
```

```python
subdf = df[['Work_accident','promotion_last_5years']]
subdf
```

```python
finaldf = pd.concat([subdf,d_satisfaction,d_hours,d_projects,d_salary,d_last_evaluation,d_time_spend_company], axis='columns')
finaldf.head()
```

```python
x = finaldf
y = df.leftOrNot
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3)
```

```python
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
model.score(x_test,y_test)
```
