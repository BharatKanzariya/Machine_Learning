# This notebook is for singlevariable LinearRegression

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
import numpy as np
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df
```

```python
%matplotlib inline
plt.scatter(df.year,df.per_capita_income,marker='+',color='red')
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(df[['year']],df.per_capita_income)
```

```python
reg.predict([[2000]])
```

```python
reg.coef_
```

```python
import sklearn.externals
import joblib
```

```python
joblib.dump(reg,'Canada_TrainModel')
```

```python
obj = joblib.load('Canada_TrainModel')
```

```python
obj.predict([[2000]])
```

```python
obj.coef_
```

```python
plt.scatter(df.year,df.per_capita_income,marker='+',color='red')
plt.plot(df.year,reg.predict(df[['year']]))
```
