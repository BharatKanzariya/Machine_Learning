```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
from word2number import w2n
```

```python
emp_data = pd.read_csv('hiring.csv')
emp_data
```

```python
emp_data.experience.fillna('Zero',inplace=True)
emp_data
```

```python
median_testscore = emp_data.test_score.median()
median_testscore
```

```python
emp_data.test_score.fillna(median_testscore,inplace=True)
emp_data
```

```python
emp_data['num_experience'] = emp_data.experience.apply(w2n.word_to_num)
emp_data
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(emp_data[['num_experience','test_score','interview_score']],emp_data.salary)
```

```python
print(reg.coef_)
print(reg.intercept_)
```

```python
reg.predict([[0,8,9]])
```

```python
df2 = emp_data[['num_experience','test_score','interview_score']]
df2
```

```python
predict_salary = reg.predict(df2)
predict_salary
```

```python
emp_data['predict_salary'] = predict_salary
```
